```python
### Import du fichier
import pandas as pd
```

```python
df_ovni = pd.read_csv(" http://bit.ly/uforeports")
df_ovni
```

```python
### 1) Quelle est la forme la plus représentée?

df_ovni['Shape Reported'].value_counts().head(1)
```

```python
### 2) Quelle est la ville où l’on a le plus de rapports de forme « FIREBALL » ?
df_ovni['City'][df_ovni['Shape Reported'] == ' FIREBALL'].value_counts().head(1)
```

```python
### 3) Quel est le pourcentage de rapports où la couleur est présente ? remplacez les valeurs de couleurs manquantes par « NONE »
```

```python
round(100 - (df_ovni['Colors Reported'].isna().sum() / len(df_ovni)* 100),2)
```

```python
df_ovni['Colors Reported'].notna().mean()*100
```

```python
df_ovni.fillna('None')
```

```python
### 4) Quelle est la décennie où l’on a recensé le plus de cas d’ovnis ?
df_ovni['Time'][0].split('/')[2][:3] + '0'
```

```python
df_ovni['Decade'] = df_ovni['Time'].apply(lambda x: x.split('/')[2][:3] + '0')
```

```python
from datetime import datetime

df_ovni['datetime_dec'] = df_ovni['Time'].apply(lambda x: datetime.strptime(x, '%m/%d/%Y %H:%M').year//10*10)
```

```python
df_ovni.head(3)
```

```python
df_ovni['Decade'].value_counts().head(1)
```

```python
### 5) Affichez le top 10 des villes où l’on voit le plus d’OVNIS en excluant New York
top_10 = df_ovni['City'][df_ovni['City'] != 'New York City'].value_counts().head(10)
```

```python
### 6) Afficher un dataframe de ces 10 villes avec les décennies, où les valeurs seront le nombre d'ovnis par combinaison
```

```python
df_TOP10 = pd.pivot_table(df_ovni, index = 'City', columns= 'Decade', values='State', aggfunc= 'count')
```

```python
df_TOP10.fillna(0, inplace = True)
```

```python
df_TOP10[df_TOP10.index.isin(top_10.index)]
```
