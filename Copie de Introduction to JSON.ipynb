# Reading a JSON

Execute the code below :

```
import json
import requests

link = "https://raw.githubusercontent.com/murpi/wilddata/master/quests/monthlySalesbyCategoryMultiple.json"

# This function gets the file content from the URL
r = requests.get(link)


# You can access to the text content with this function
r.text
```

```
# Try this
import json

data = json.loads( r.text)
data
```

# Manipulate the data

Now print for each category their total number of sales (without using a dataframe, just manipulating the dictionaries and lists).

Expected result:
- Furniture 568
- Technology 676



```
category_sales = {}
for number_sale in data['contents']:
  #print(number_sale)
  category = number_sale['category']
  #print(category)
  monthlySales = number_sale['monthlySales']
  #print(monthlySales)
  total_sales = sum(sale['sales'] for sale in monthlySales)
  #print(f'Furniture : {total_sales}, Technology:{total_sales}')
  category_sales[category] = total_sales


for category, total_sales in category_sales.items():
   print(f'{category} {total_sales}')


```
