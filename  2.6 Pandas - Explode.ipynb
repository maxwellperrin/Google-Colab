- Which actor is mentioned the most often?


```
import pandas as pd
import os

link = 'https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/imdb_1000.csv'
df = pd.read_csv(link)
df
```

```
df['actors_list'] = df['actors_list'].map(str)
df['actors_list']
```

```
df['actors']=df['actors_list'].apply(lambda x: eval(x))
df['actors']
```

```
df
liste = []

for acteurs in df['actors_list'] :
  acteurs = eval(acteurs)
  liste.append(acteurs)

df['actors_list'] = liste
```

```
df.explode('actors')
```

```
df_1 = df['actors'].explode()
```
