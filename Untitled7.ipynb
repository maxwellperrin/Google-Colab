```python
df = pd.read_csv(r'https://github.com/murpi/wilddata/raw/master/quests/tweets.zip')
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
nltk.download('punkt')
nltk.download('wordnet')
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
nltk.download('wordnet')
nltk.download('stopwords')
```

```python
def clean(text):
    tokens = word_tokenize(text)

    # Initialisation du lemmatizer
    lemmatizer = WordNetLemmatizer()

    # Lemmatisation de chaque token
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]

    # Retourne les tokens lemmatisés sous forme de chaîne de caractères
    return ' '.join(lemmatized_tokens)

# Test de la fonction
sentence = "You are better when I am well."
result = clean(sentence)
print(result)
```

```python
def clean_with_stemmer(sentence):
    # Tokenisation de la phrase
    tokens = word_tokenize(sentence)

    # Initialisation du stemmer
    stemmer = PorterStemmer()

    # Application du stemmer à chaque token
    stemmed_tokens = [stemmer.stem(token) for token in tokens]

    # Retourne les tokens stemmés sous forme de chaîne de caractères
    return ' '.join(stemmed_tokens)

# Test de la fonction
sentence = "You are better when I am well."
result = clean_with_stemmer(sentence)
print(result)
```

```python
df['clean']
```

```python
from sklearn.model_selection import train_test_split




X = df['clean']


y = df['sentiment']


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=32)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report




vectorizer = CountVectorizer()


X_train_vectorized = vectorizer.fit_transform(X_train)
X_test_vectorized = vectorizer.transform(X_test)

classifier = MultinomialNB()
classifier.fit(X_train_vectorized, y_train)


predictions = classifier.predict(X_test_vectorized)

# Évaluation du modèle
accuracy = accuracy_score(y_test, predictions)
report = classification_report(y_test, predictions)

print(f"Accuracy: {accuracy}")
print("Classification Report:")
print(report)
```
