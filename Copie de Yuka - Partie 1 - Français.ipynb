# Mission 4

Quel est le nombre de colonne appartenant à chaque quartile des taux de remplissage ? Autrement dit, combien de colonne sont remplies à 0%, entre 0 et 25%, entre 25 et 50%...

```
def remplissage(dataframe):
  taux_de_remplissage = (dataframe.count()/len(dataframe))* 100

  quartiles = {
        '0-25%': 0,
        '25-50%': 0,
        '50-75%': 0,
        '75-100%': 0
    }

  for taux in taux_de_remplissage:
        if taux >= 0 and taux <= 25:
            quartiles['0-25%'] += 1
        elif taux > 25 and taux <= 50:
            quartiles['25-50%'] += 1
        elif taux > 50 and taux <= 75:
            quartiles['50-75%'] += 1
        elif taux > 75 and taux <= 100:
            quartiles['75-100%'] += 1
  return quartiles

remplissage(df)
```
