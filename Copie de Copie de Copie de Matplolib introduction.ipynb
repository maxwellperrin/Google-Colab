# Execute the code below

```
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import datetime as dt

link = "https://raw.githubusercontent.com/murpi/wilddata/master/quests/weather2019.csv"
df_weather = pd.read_csv(link).iloc[:,[0, 1, 2, -3]]

df_weather['DATE'] = pd.to_datetime(df_weather['DATE'])
```

# Now, complete the following tasks :

## ScatterPlot with 2 variables, title and legend
First of all, you have to create a scatterplot with MAX (in red) and MIN (in blue) Temperatures on the same graph

X-axis stay "DATE" for both values

You have to add title and legend

Here is an example of what is expected :

matplotlib1.png

```
# Complete the code below :
plt.title(
    "MIN and Max Temperatures(AMERICA GREAT AGAIN)",
    fontdict={"family": "serif", "color": "black", "weight": "bold", "size": 18},
)


plt.scatter(x = df_weather["DATE"], y = df_weather["MAX_TEMPERATURE_C"], c= 'Red', s = 10, marker = "*")
plt.scatter(x = df_weather["DATE"], y = df_weather["MIN_TEMPERATURE_C"], c= 'blue' , s = 10)

plt.ylabel('Températures')

temperature =['Max', 'Min']
plt.legend(temperature,
           loc='upper left',
           fontsize = 6,
           title = 'temperatures',
           title_fontsize = 6,
           shadow = True,
           fancybox = True,
           facecolor = 'lightcyan',
           edgecolor = 'green',
           markerscale = 0.6

           )


```

## 2 charts on the same figure
Now, complete this code to have MAX and MIN Temperature in 2 charts, MAX on the left side and MIN on the right side

The X axis have to be the date of the year

Each graph must have a title, a legend for the colors, and of course, readable axis (ticks)

Here is an example of what is expected :


Capture d’écran 2021-08-26 à 17.27.31.png

```
# Complete the code below :

fig, (graph1, graph2) = plt.subplots(1, 2, figsize=(12, 4))


plt.ylabel('Températures Max')
graph1.scatter(x = df_weather["DATE"], y = df_weather["MAX_TEMPERATURE_C"], c= 'Red', s = 10, marker = "*")
graph1.yaxis.set_ticks(range(-5, 41, 5)) # définis l'axe des ordonnées(y)
graph1.set_title(
    " Max Temperatures",
    fontdict={"family": "serif", "color": "black", "weight": "bold", "size": 10},
)
graph1.set_ylabel('Temperature MAX')





graph2.scatter(x = df_weather["DATE"], y = df_weather["MIN_TEMPERATURE_C"], c= 'blue' , s = 10, marker ="*")
graph2.yaxis.set_ticks(range(-5, 41, 5))
graph2.set_title(
    " Min Temperatures",
    fontdict={"family": "serif", "color": "black", "weight": "bold", "size": 10},
)


graph2.set_ylabel('Temperature MIN') # donne un titre à l'axe des ordonnées (y)












plt.show()
```
